```python
"""
'A Blacksmith' by Benjamin G. Garbinski.
It was made on the weekend of 23-24 in October 2021 for a python class.

The story is simple, and it is not based on anything specifically but more on 
the concept of an evil ruler abusing his power over a simple blacksmith. A 
trope that has been used many times. The player does not need any pre-existing
knowledge, but to understand some terms, a shallow understanding of metalwork 
could help. The "rooms" or "stages" in this game are pretty close to one another 
as the whole story unfolds in a poor blacksmiths workshop. I have added some D&D 
rolls to spice things up.

ver: 0.2

Known bugs: 0
"""
#importing exit so, that at the fail or win condition, the game can be exited.
from sys import exit
#importing random to use it as a spice for extra details in the story and enhance
#replayability.
import random

#first of all "stages" introduces the story and presents the characters.
def shaping():
    
    input(prompt='<Press enter to begin the story>')
    
#roll to see if the player gets more flavor text in the story or not.    
    perception_start = random.randint(1, 2)

    if perception_start == 2:
        print(""" 
You are a blacksmith making a blade for your lord. He is impatiently waiting 
in the storefront, and he is NOT known for leniency or patience, but seems 
distracted by a wench that accompanies him.

<perception check successful>

You know the wench; it is rumored in the village that the lord fancies her. As 
you remember this, the wench winks at you.
    """)
        
    else:
        print("""
You are a blacksmith making a blade for your lord. He is impatiently waiting 
in the storefront, and he is NOT known for leniency or patience, but seems 
distracted by a wench that accompanies him.

<perception check failed>
    """)

#First actual choice, the player should understand that something terrible will 
#happen if they tell the lord to come back, given the emphasis on the "not" in
#the introduction.    
    print("""
1) Tell him the blade is not ready and to come back tomorrow.
2) Quickly thank the gods for the wench and escape to your workshop to start 
hammering.
    """)
#record input from the player.   
    wait_lord = input(prompt="> ")
    
#To normalize responses and "Chase" away bugs.
    wait_lord = wait_lord.casefold()
    
#Some "or" to catch some common words in case the player does not feel like
#using numbers.
    if wait_lord == '1' or "tomorrow" in wait_lord or "not ready" in wait_lord:
        
        print("""
You go to the storefront and tell him to come back tomorrow.
    """)
        
#First way for your blacksmith to die also calling the function to fail the 
#player.        
        fail()

#Again, some "or" to catch some common words used. This is the winning path with
#another fork the total capitalization should indicate panic, which would lead to 
#mistakes, if the dirty part was not a strong enough clue that it's a wrong choice.    
    elif wait_lord == "2" or "start" in wait_lord or "hammer" in wait_lord:
        
#more spice for the spice lords!
        perception_door = random.randint(1, 2)

        if perception_door == 2:
            print("""
<perception check successful>

You hear the door to the storefront opening. Your lord is standing there holding 
hands with the wench, waiting to see you begin hammering.

You are glad to have some hot steel ready for hammering. Should you start 
making a high-quality piece, or something fast and dirty?

1)It's my lord; quality is the only way to keep my head.
2)HE IS WAITING; HAMMER SOMETHING FAST!
        """)
        else:
            print("""
<perception check failed>

Should you start making a high-quality piece, or something fast and dirty?

1)It's my lord; quality is the only way to keep my head.
2)HE IS WAITING; HAMMER SOMETHING FAST!
        """)
        
        quality = input(prompt="""> """)
#To normalise responses and "Chase" away bugs.
        quality = quality.casefold()
    
#Again, some "or" to catch some common words used.        
        if quality == "1" or "quality" in quality:

#Add some extra spice to flesh out the wench character as the blacksmith's guardian 
#angel
            if perception_door == 2:
                print("""
You hammer and hammer with precision and enthusiasm like your life depended on 
it, and by the end, you have in your hands the beginning of a beautiful blade.

-Wench: See, I told you it would be worth the wait. 

The wench points at the blade you just hammered into shape, to which the lord 
just grumbles and nods in agreement.

-Wench: Let us leave the blacksmith to his forge; this heat is not good for my 
skin.

As the wench pulls the lord by the hand back to the storefront, you make a 
mental note to buy her anything she wants from the traveling merchant next time 
he stops by the village.
            """)
                
            else:
                print("""
You hammer and hammer with precision and enthusiasm like your life depended on 
it, and by the end, you have in your hands the beginning of a beautiful blade.
            """)

            sanding()
            
#Again, some "or" to catch some common words used.        
        elif quality == "2" or "fast" in quality:
            
            print(""" 
You hammer like a man possessed by the devil, panic gripping you tightly. By 
the end, the blade is twisted. You do not know when he came in, but you hear 
your lord speak from behind.
            
-Lord: So that is what you think your "twisted" lord deserves; a twisted blade.
            """)

#Second fail condition            
            fail()
            
#This is the catch-all from the first "if" of the nested. Also, a reference to
#Ralph from Simpsons.            
        else:
            print("""
Something went wrong in 'quality'. Hehe! I'm in danger!\n
                """)
            input(prompt='<Press any key to restart>\n')
            shaping()
            
#This is the catch-all from the first "if" of the nested. Also, a reference to
#Ralph from Simpsons.
    else:
        print("""
Something went wrong in 'wait_lord'. Hehe! I'm in danger!\n
        """)
        
        input(prompt='<Press any enter to restart>\n')
        
        shaping()
    

#Second stage comprehends the reduction of excess metal from the shaped ingot.
def sanding():
    print("""
The body of the blade is ready, now to sand it into a better shape.
    """)
#Create a base to reduce.    
    thickness = 6

#While to 1 less than required to progress as to break it even if "ifs" fail.     
    while thickness > 2:

#Set the wanted "thickness" and condition to progress.        
        if thickness == 3:
            print("""
The blade is ready for quenching! You quickly put it in the forge to heat up.
                """)
        
            quenching()
            
#Sanding process 1 by 1 with small comment.
        elif thickness >= 4:
            input(prompt='<Press enter to sand it>')

            print("""
The metal sparks and is reduced!
                """)
        
            thickness -= 1
            
#Here comes Ralph!    
        else:
            print("""
Something went wrong in 'thickness'. Hehe! I'm in danger!\n
                """)
        
            input(prompt='<Press enter to restart>\n')
        
            sanding()            

#Third stage in which the blacksmith is almost finished, but I added a nice trap 
#for those that in a "hurry situation" would stall to see what happens.
def quenching():
    
    print("""
Sweat covers your whole body, heat and stress are not a good combination, but
you can almost see the end of this ordeal. Now just 15 seconds or more of 
quenching, and the blade will be ready.
        """)
    
    time_quench = input(prompt="""
With a red-hot blade at the ready; how many seconds do you submerge the blade 
in oil?
                        """)
    
#A try\except to "Chase" bugs away if the player puts a string instead
#of a number and returning them to try again.
    try:

        time_quench = int(time_quench)
        
    except:
        print("""
That is not a time in round seconds, like 1 or 100. Please try again.
        """)
        quenching()
        
#Condition for success capturing 15 to 60 including.        
    if time_quench >= 15 and time_quench <= 60:
        
        print("""
*TSSS!!!* The oil boils and roils. After it calms down, you take the blade out
and with a smile, get ready to deliver.
            """)
        
        finishing()

#Prematurely removing it leads to premature death for our blacksmith.         
    elif time_quench < 15:
        
        print("""
*TSSS!!* the oil boils! and you take out the blade too early. *CREAK* the 
blade cracks, and the color fades from your face. You do not have time to forge 
another. Resigned, you drag yourself to the storefront, broken blade in 
hand, to inform the lord that he will have to come back later.

-Lord: A broken blade from a broken blacksmith. Fitting.
            """)
        
        fail()
        
#Trap for stallers.        
    elif time_quench > 60:
        
        print("""
You start nodding while watching the blade quench as tiredness and relief 
wash over you. Long after the oil stops reacting ,you lose your fight 
to stay conscious and fall asleep. You are woken up by your lord, irritation
clear as day in his face.

Lord: Serving your lord is that boring, is it blacksmith?
            """)

        fail()

#Why hello there, Ralph!        
    else:
        
        print("""
Something went wrong in 'time_quench'. Hehe! I'm in danger!\n
            """)
        
        input(prompt='<Press any key to restart>\n')

        quenching()

#Fourth and final stage of our game in which he almost loses his head over 
#haste but manages to honor his liege properly.
def finishing():
    
    wench_luck = random.randint(1, 10)
    
    print("""
Just as you were about to leave the workshop, you remembered you needed to put 
a hilt with his name and the kingdom's name in it. This is not some peasants' 
blade dolt!
    """)
    
    lord_name = input(prompt="""
What is your lordship name?    
    """)

#Names should be capitalized.    
    lord_name = lord_name.capitalize()
    
    print(f"""
You quickly hammer the hilt in and start carving: Lord {lord_name} of... of...
        """)

#Names of kingdoms are still names    
    kingdom_name = input(prompt="""
What is this kingdom name?    
    """)
    
    kingdom_name = kingdom_name.capitalize()
    
#Some final spice for players to hunt the last "bad end".
    if wench_luck > 1:
        print(f"""

Yes! Yes, it's all coming together: 'His grace lord {lord_name} of the great 
{kingdom_name}', is carved in beautiful italics on the hilt. You wrap the now 
complete sword and bolt to the storefront.

<You got lucky!>

When the door opens, lord {lord_name} is waiting with a smile, and the wench is 
nowhere to be seen. Lucky you! He seems to be in a good mood and immediately  
accepts the sword without even inspecting it!
        """)
    
        victory()

    else:
        print(f"""
Yes! Yes, it's all coming together: 'His grace lord {lord_name} of the great 
{kingdom_name}', is carved in beautiful italics on the hilt. You wrap the now 
complete sword and bolt to the storefront.

<You got unlucky!>

Your premature celebration ends as soon as you enter the storefront. The scowl
in lord {lord_name} face says everything you need to know. It seems the wench
could not buy you more time.
            """)
        fail()
        
#why not give our player a little fanfare when they won.          
def victory():
    
    print("""
Congratulations! You managed to survive the lord and win!
           """)
    
    input(prompt='<Press any key to exit the game>\n')
    
    exit()

#Here are the failures at this game, with the option to try again   
def fail():
    
    print("""
Oof. The lord did not like that, off with your head for betraying the kingdom.

Try again?
1) Yes.
2) No.
        """)
    
    done = input(prompt="""> """)
    done = done.casefold()
    
    if done == "1" or "yes" in done:
        shaping()
        
    elif  done == "2" or "no" in done:
        exit()

#calling to start the game.
shaping()
```
